```python
import apss_lib
apss_lib.verbose=False
import numpy as np
```

```python
for sol in np.arange(400):
    try:
        dataw = apss_lib.getsol(sol,distant=True,var="HWS")
        fq = dataw['WIND_FREQUENCY']
        print sol, fq[0:3]
    except:
        pass
```

```python
for sol in np.arange(400):
    try:
        datap = apss_lib.getsol(sol,distant=True,var="PRE")
        print sol, np.mean(datap['PRESSURE_FREQUENCY'])
    except:
        pass
```
