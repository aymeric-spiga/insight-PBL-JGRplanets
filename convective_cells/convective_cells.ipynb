```python
import numpy as np
import apss_lib 
%matplotlib inline
```

```python
soltab = np.arange(234,240,1)
soltab = np.arange(234,236,1)
vartab = ["HWS","PRE","WD","MAT"]
soltab = np.arange(234,235,1)   
```

```python
mint = 8.5
maxt = 17.5
mint = 11 #10.0
maxt = 17 


for sol in soltab:
    
    datap = apss_lib.getsol(sol,var="PRE",distant=True)
    dataw = apss_lib.getsol(sol,var="HWS",distant=True)
    
    for code in vartab:
        freq = 1.
        addsmooth=True
        zedata = dataw
        ymin = ymax = None
        if code == "PRE":
            freq = 10.
            addsmooth = True #False
            zedata = datap
        if code == "WD":
            ymin = 50
            ymax = 180 
        apss_lib.do_wavelet(zedata,code=code,freq=freq,\
                            filename=code,\
                            ymin=ymin,ymax=ymax,\
                            title="",\
                            window=5000,addsmooth=addsmooth,\
                            mint=mint,maxt=maxt,timetype='LTST',\
                            tmin=100.,tmax=5000.)
```
