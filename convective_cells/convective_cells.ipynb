```python
soltab = np.arange(234,240,1)
soltab = np.arange(234,236,1)
soltab = np.arange(234,235,1)
datatab = [apss_lib.getsol(sol) for sol in soltab]    
```

```python
#for data in datatab:
#    apss_lib.showfreq(data)
```

```python
for data in datatab:
    for code in ["MAT","HWS","PRE","WD"]:
        freq = 1.
        if code == "PRE":
            freq = 10.
        apss_lib.do_wavelet(data,code=code,freq=freq,\
                            filename=code,\
                            window=5000,addsmooth=True,\
                            mint=8.5,maxt=17.5,timetype='LMST',\
                            tmin=100.,tmax=5000.)
```
