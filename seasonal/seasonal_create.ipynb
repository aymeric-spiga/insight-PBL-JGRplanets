```python
import apss_lib
import numpy as np
%matplotlib inline
apss_lib.verbose = False
```

```python
solini = 0 ; solsol = 400 ; ttinter = [[11,15]]
ttinter = [[12,13]]
ttinter = [[12,14]]
ttinter = [[11,13]]
ttinter = [[10,15]]
ttinter = [[11,14]]
```

```python
pre,prex = apss_lib.multisol(pds=True,solini=solini,solsol=solsol,code="PRE",ttinter=ttinter,\
                        timetype="LTST",\
                        compute="mean",ymin=500,ymax=800,\
                        addls=True,addlspos=1.03,\
                        fmt="%.1f",marker="o",nxticks=20,facselec=0.95)
```

```python
np.save("pre.npy",pre)
np.save("prex.npy",prex)
```

```python
datext = "tsurf_bins_LT_%i_%i.txt"%(ttinter[0][0],ttinter[0][1])
pat,patx = apss_lib.multisol(pds=True,solini=solini,solsol=solsol,code="PAT",ttinter=ttinter,\
                        timetype="LTST",\
                        compute="mean",ymin=230,ymax=290,\
                        addls=True,addlspos=1.008,\
                        fmt="%.1f",marker="o",nxticks=20,facselec=0.95,\
                        addtxt=datext)
mat,matx = apss_lib.multisol(pds=True,solini=solini,solsol=solsol,code="MAT",ttinter=ttinter,\
                        timetype="LTST",\
                        compute="mean",ymin=230,ymax=290,\
                        addls=True,addlspos=1.008,\
                        fmt="%.1f",marker="o",nxticks=20,facselec=0.95,\
                        addtxt=datext)
```

```python
np.save("pat.npy",pat)
np.save("patx.npy",patx)
np.save("mat.npy",pat)
np.save("matx.npy",patx)
```

```python
hws,hwsx = apss_lib.multisol(pds=True,solini=solini,solsol=solsol,code="HWS",ttinter=ttinter,\
                        timetype="LTST",\
                        compute="mean",ymin=2,ymax=12,\
                        addls=True,addlspos=1.03,\
                        fmt="%.1f",marker="o",nxticks=20,facselec=0.95)
hwsstd,hwsstdx = apss_lib.multisol(pds=True,solini=solini,solsol=solsol,code="HWS",ttinter=ttinter,\
                        timetype="LTST",\
                        compute="std",ymin=1,ymax=4,\
                        addls=True,addlspos=1.06,\
                        fmt="%.1f",marker="o",nxticks=11,facselec=0.95)
```

```python
np.save("hws.npy",hws)
np.save("hwsx.npy",hwsx)
np.save("hwsstd.npy",hwsstd)
np.save("hwsstdx.npy",hwsstdx)
```

```python
hwd,hwdx = apss_lib.multisol(pds=True,solini=solini,solsol=solsol,code="WD",ttinter=ttinter,\
                        timetype="LTST",\
                        compute="mean", ymin=-180, ymax=180,\
                        addls=True,addlspos=1.03,\
                        fmt="%.1f",marker="o",nxticks=20,facselec=0.95)
```

```python
np.save("hwd.npy",hwd)
np.save("hwdx.npy",hwdx)
```
