```python
tmin,tmax = 230,290
tmin,tmax = 150,300
datext = "tsurf_bins_LT_%i_%i.txt"%(ttinter[0][0],ttinter[0][1])
pat,patx = apss_lib.multisol(pds=True,solini=solini,solsol=solsol,code="PAT",ttinter=ttinter,\
                        timetype="LTST",\
                        compute="mean",ymin=tmin,ymax=tmax,\
                        addls=True,addlspos=1.008,\
                        fmt="%.1f",marker="o",nxticks=20,facselec=0.95,\
                        addtxt=datext)
mat,matx = apss_lib.multisol(pds=True,solini=solini,solsol=solsol,code="MAT",ttinter=ttinter,\
                        timetype="LTST",\
                        compute="mean",ymin=tmin,ymax=tmax,\
                        addls=True,addlspos=1.008,\
                        fmt="%.1f",marker="o",nxticks=20,facselec=0.95,\
                        addtxt=datext)
```

```python
np.save("pat_%i_%i.npy"%(ttinter[0][0],ttinter[0][1]),pat)
np.save("patx_%i_%i.npy"%(ttinter[0][0],ttinter[0][1]),patx)
np.save("mat_%i_%i.npy"%(ttinter[0][0],ttinter[0][1]),mat)
np.save("matx_%i_%i.npy"%(ttinter[0][0],ttinter[0][1]),matx)
```
