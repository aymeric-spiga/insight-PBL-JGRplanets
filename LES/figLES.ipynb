```python
import apss_lib
import numpy as np
import ppplot
%matplotlib inline
```

```python
fold = "./"
```

```python
lim = "0p3"
#lim = "0p5"

fifi = "/txt_sol_to_sol/solDD_above_"+lim+".txt"

one = fold+"25M_LS300_W20"+fifi
two = fold+"25M_LS0_W10"+fifi
#two = fold+"25M_LS300_W10"+fifi
#tre = fold+"25M_LS120_W20"+fifi
tre = fold+"25M_LS30_W20"+fifi
qua = fold+"25M_LS120_W20"+fifi

#one = fold+"25M_LS300_W20"+fifi
#two = fold+"25M_LS30_W20"+fifi
#tre = fold+"25M_LS120_W20"+fifi

config = [(fold+"25M_LS300_W20"+fifi,\
           fold+"25M_LS0_W10"+fifi,\
           fold+"25M_LS30_W20"+fifi,\
           fold+"25M_LS120_W20"+fifi),\
          (fold+"25M_LS300_W10"+fifi,\
           fold+"25M_LS0_W10"+fifi,\
           fold+"25M_LS30_W10"+fifi,\
           fold+"25M_LS120_W10"+fifi)]
```

```python
count = 1

for ccc in config:

    foo,foo,oned,foo = np.loadtxt(ccc[0],unpack=True,skiprows=1)
    foo,foo,twod,foo = np.loadtxt(ccc[1],unpack=True,skiprows=1)
    foo,foo,tred,foo = np.loadtxt(ccc[2],unpack=True,skiprows=1)
    foo,foo,quad,foo = np.loadtxt(ccc[3],unpack=True,skiprows=1)
   
    onei = np.arange(0,41)
    twoi = np.arange(41,151)
    trei = np.arange(151,321)
    quai = np.arange(321,401)

    onedd = oned[onei]
    twodd = twod[twoi]
    tredd = tred[trei]
    quadd = quad[quai]
    
    ssol = 0
    esol = 400

    ppplot.changefont(16)
    ppplot.changemarkersize(7)

    nx = 16 #32 #24 #16
    ny = 4 #6
    fig = ppplot.figuref(x=nx,y=ny)
    pl = ppplot.plot1d(fig=fig)
    pl.xmin = ssol
    pl.xmax = esol+1      
    pl.nxticks = 10
    pl.ymin = 0
    pl.ymax = 40 #60
    #pl.ymax = 15
    pl.nyticks = np.min([pl.ymax - pl.ymin + 1, 10])
    pl.fmt = '%.0f'
    pl.xlabel = "Large Eddy Simulations InSight-equivalent sol"
    pl.ylabel = r'# pressure drops above ' + lim.replace("p",".").replace("_"," ") + " Pa" #per hour'
    pl.marker = "s"
    pl.linestyle = ""

    #pl.f = alld

    pl.f = onedd
    pl.x = onei
    pl.make()

    pl.f = twodd
    pl.x = twoi
    pl.make()

    pl.f = tredd
    pl.x = trei
    pl.make()

    pl.f = quadd
    pl.x = quai
    pl.make()

    apss_lib.axisls(pl,addlspos=1.05,col="grey")

    #ppplot.show()
    ppplot.save(mode="pdf",filename="season_les_%i" % (count))
    
    count += 1




```
